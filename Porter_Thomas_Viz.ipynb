```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import unitary_group
```

```python

# Parameters
num_qubits = 9  # Number of qubits
dim = 2**num_qubits  # Hilbert space dimension
```

```python

# Step 1: Generate a random unitary matrix
U = unitary_group.rvs(dim)

# Step 2: Prepare the initial state |0‚ü©
psi = np.zeros(dim)
psi[0] = 1  # All-zero state
```

```python

# Step 3: Apply the unitary transformation
phi = U @ psi

# Step 4: Compute measurement probabilities
probabilities = np.abs(phi)**2

# Step 5: Normalize and collect Porter-Thomas data
normalized_probabilities = dim * probabilities  # Scale probabilities by dimension

# Step 6: Plot the histogram and Porter-Thomas distribution
plt.hist(normalized_probabilities, bins=50, density=True, label="Simulated Data", alpha=0.75)
x = np.linspace(0, 5, 1000)
plt.plot(x, np.exp(-x), label="Porter-Thomas Distribution (e^(-x))", color="red")
plt.title("Porter-Thomas Distribution for a Random Quantum State")
plt.xlabel("Normalized Probability (N * p)")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
